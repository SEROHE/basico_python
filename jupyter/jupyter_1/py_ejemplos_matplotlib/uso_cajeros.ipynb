```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv('943.csv')

# Elimina la columna 'AO'
df = df.drop('AO', axis=1)

# Establece 'INSTALACION' como ndice
dfi = df.set_index('INSTALACION')

# Agrupa por 'SERVICIO' y suma los usos
tabla = dfi.groupby(by='INSTALACION')['USOS'].sum()

# Crea el grfico de barras horizontales
fig, ax = plt.subplots()

# Obtiene las etiquetas de los servicios
x_pos = range(len(tabla))

# Crea las barras horizontales
hbars = ax.bar(x_pos, tabla, align='center')

# Etiqueta el eje x con los servicios
ax.set_xticks(x_pos)
ax.set_xticklabels(tabla.index)

# Aade etiquetas en las barras
ax.bar_label(hbars, fmt='%.0f')

ax.tick_params(axis='x', rotation=90)


# Ajusta el espacio entre las barras
plt.tight_layout()

# Muestra el grfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv('943.csv')

# Elimina la columna 'AO'
df = df.drop('AO', axis=1)

# Agrupa por 'INSTALACION' y 'SERVICIO' y suma los usos
tabla = df.groupby(['INSTALACION', 'SERVICIO'])['USOS'].sum().unstack()

# Configura el grfico
plt.figure(figsize=(12, 8))

# Itera sobre las instalaciones
for i, (instalacion, usos) in enumerate(tabla.iterrows(), start=1):
    plt.bar([i + 0.2 * j for j in range(len(usos))], usos, width=0.2, label=instalacion, align='center')

# Configura etiquetas y leyenda
plt.xlabel('Servicio')
plt.ylabel('Cantidad de usos')
plt.title('Cantidad de usos por servicio para cada instalacin')
plt.xticks(range(1, len(tabla.columns) + 1), tabla.columns, rotation=90)
plt.legend(title='Instalacin')


# Muestra el grfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv('943.csv')

# Elimina la columna 'AO'
df = df.drop('AO', axis=1)

# Agrupa por 'INSTALACION' y 'SERVICIO' y suma los usos
tabla = df.groupby(['INSTALACION', 'SERVICIO'])['USOS'].sum().unstack()

# Configura el grfico
plt.figure(figsize=(12, 8))

# Itera sobre las columnas de la tabla (los servicios)
for servicio in tabla.columns:
    plt.bar(tabla.index, tabla[servicio], label=servicio)

# Configura etiquetas y leyenda
plt.xlabel('Instalacin')
plt.ylabel('Cantidad de usos')
plt.title('Cantidad de usos por instalacin para cada servicio')
plt.xticks(rotation=90)
plt.legend(title='Servicio', bbox_to_anchor=(1.05, 1), loc='upper left')

# Muestra el grfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
df = pd.read_csv('943.csv')

# Elimina la columna 'AO'
df = df.drop('AO', axis=1)

# Obtener la lista de instalaciones
instalaciones = df['INSTALACION'].unique()

# Iterar sobre las instalaciones y generar un grfico para cada una
for instalacion in instalaciones:
    # Filtrar los datos para la instalacin actual
    datos_instalacion = df[df['INSTALACION'] == instalacion]
    
    # Configurar el grfico
    plt.figure(figsize=(12, 8))
    
    # Iterar sobre los servicios y generar las barras para esta instalacin
    for servicio in datos_instalacion['SERVICIO'].unique():
        datos_servicio = datos_instalacion[datos_instalacion['SERVICIO'] == servicio]
        plt.bar(datos_servicio['SERVICIO'], datos_servicio['USOS'], label=servicio)
    
    # Configurar etiquetas y leyenda
    plt.xlabel('Servicio')
    plt.ylabel('Cantidad de usos')
    plt.title(f'Cantidad de usos por servicio para la instalacin {instalacion}')
    plt.xticks(rotation=90)
    plt.legend(title='Servicio', bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # Mostrar el grfico
    plt.tight_layout()
    plt.show()
```
