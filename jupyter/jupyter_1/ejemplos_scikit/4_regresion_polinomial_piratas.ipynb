```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression, LogisticRegression

# Datos ficticios para ilustrar la relación entre temperatura y cantidad de piratas
temperaturas = np.array([20, 23, 24, 25, 28]).reshape(-1, 1)  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Crear un objeto PolynomialFeatures para generar características polinomiales
grado_polinomio = 2  # Grado del polinomio deseado
polinomio_features = PolynomialFeatures(degree=grado_polinomio)

# Generar características polinomiales para las temperaturas
temperaturas_polinomiales = polinomio_features.fit_transform(temperaturas)

# Crear un modelo de regresión lineal
modelo = LogisticRegression()

# Entrenar el modelo con las características polinomiales y la cantidad de piratas
modelo.fit(temperaturas_polinomiales, cantidad_piratas)

# Generar un rango de valores de temperatura para la curva de regresión
temperaturas_prediccion = np.linspace(min(temperaturas), max(temperaturas), 100).reshape(-1, 1)

# Generar características polinomiales para los valores de temperatura de predicción
temperaturas_prediccion_polinomiales = polinomio_features.transform(temperaturas_prediccion)

# Predecir la cantidad de piratas para los valores de temperatura de predicción
cantidad_piratas_prediccion = modelo.predict(temperaturas_prediccion_polinomiales)

# Graficar los datos originales y la curva de regresión polinomial
plt.scatter(temperaturas, cantidad_piratas, color='blue', label='Datos Originales')
plt.plot(temperaturas_prediccion, cantidad_piratas_prediccion, color='red', label=f'Regresión Polinomial (Grado {grado_polinomio})')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Cantidad de Piratas')
plt.title('Relación entre Temperatura y Cantidad de Piratas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos ficticios para ilustrar la relación entre temperatura y cantidad de piratas
temperaturas = np.array([20, 23, 24, 25, 28])  # Ejemplo de temperatura (en grados Celsius)
cantidad_piratas = np.array([50, 40, 30, 20, 10])  # Ejemplo de cantidad de piratas

# Ajustar un polinomio de grado 2 a los datos
coefs = np.polyfit(temperaturas, cantidad_piratas, 1)

# Crear el polinomio a partir de los coeficientes
polinomio = np.poly1d(coefs)

# Generar valores de temperatura para la curva de regresión
temperaturas_prediccion = np.linspace(min(temperaturas), max(temperaturas), 10)

# Evaluar el polinomio en los valores de temperatura de predicción
cantidad_piratas_prediccion = polinomio(temperaturas_prediccion)

# Limitar la cantidad de piratas máxima a 50
cantidad_piratas_prediccion = np.minimum(cantidad_piratas_prediccion, 50)

# Graficar los datos originales y la curva de regresión polinomial
plt.scatter(temperaturas, cantidad_piratas, color='blue', label='Datos Originales')
plt.plot(temperaturas_prediccion, cantidad_piratas_prediccion, color='red', label='Regresión Polinomial')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Cantidad de Piratas')
plt.title('Relación entre Temperatura y Cantidad de Piratas')
plt.legend()
plt.grid(True)
plt.show()
```
