```python
df = pd.DataFrame({
    'A': [1, 1, 2, 3, 3],
    'B': ['apple', 'grape', 'apple', 'banana', 'grape'],
    'C': [1, 2, 3, 4, 5]
})
print(df)

df = df.drop_duplicates(['A'])
print(df)

df = df.drop_duplicates(['B'])
print(df)

df.reset_index(drop=True, inplace=True)
print(df)
```

```python
import pandas as pd

fruit_price = r"""product,price
au-apple-1,2.5
au-apple-2,5.356
au-peach-1,4.99
au-peach-2,3.99
au-grape-2,3.99
"""

df = pd.read_csv(StringIO(fruit_price))
df['Is Apple'] = df['product'].apply(lambda x: 'Y' if 'apple' in x else 'N')

print(df)
```

```python
import pandas as pd
from io import StringIO

data = r"""date,col_a,col_b,col_c
2018-05-20,1,2,3
2018-05-21,3,4,6
2018-05-22,5,6,9
"""

df = pd.read_csv(
    StringIO(data),
    index_col=['date'],
    parse_dates=True,
    
)
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')

assert isinstance(df.index, pd.core.indexes.datetimes.DatetimeIndex)
print(df)

df.drop(['col_a'], axis=1, inplace=True)
print(df)

df.drop(['col_b', 'col_c'], axis=1, inplace=True)
assert df.empty is True, "Expecting an empty data frame"
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    data=np.random.randint(0, 10, (5, 3)),
    columns=list(['C', 'A', 'B']),
    index=[11, 22, 2, 12, 4]
)

df.reset_index(inplace=True)
print(df)

# sort the rows
df_sorted = df.sort_index(ascending=False)
print(df_sorted)

# sort the columns
df_sorted = df_sorted.sort_index(axis=1)
print(df_sorted)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO


data = r"""
date,col_num
2018-03-01,1
2018-03-02,2
2018-03-03,3
2018-03-04,4
2018-03-05,5
2018-03-06,6
2018-03-07,7
"""

df = pd.read_csv(StringIO(data),
                 index_col=['date'],
                 parse_dates=True,
                 
                 )
df.index = pd.to_datetime(df.index, format='%Y-%m-%d')

df['lag_1'] = df['col_num'].shift(1)
df['lag_2'] = df['col_num'].shift(2)
df['lag_-1'] = df['col_num'].shift(-1)
print(df)

# show
df.plot(figsize=(10, 6), marker='o')
plt.show()
```
